```python
import pandas as pd
import numpy as np
from openpyxl import load_workbook
import re
import os
import time
```

# Reading Files

```python
def valor_str(v):
  return(float(v.value.replace('.','').replace('R$ ','').replace(',','.')))
```

```python
path = os.getcwd()
```

```python
#seleciona xlsx files
spreadsheets = [f for f in os.listdir(path) if f.endswith('.xlsx')]
spreadsheets
```

```python
def read_xlsx_xp_df(ws):
  df = pd.DataFrame(columns =['tipo','ativo','valor'])
  d = []
  row = 0
  for i in ws['A']:
      row = row + 1
      color = i.fill.start_color.value
      if color == '00f0f0f0':
        tipo = i.value
        delimiter = row
        d.append({tipo:delimiter})
  for i in range(len(d)-1):
    lower = list(d[i].items())[0][1] 
    upper =  list(d[i+1].items())[0][1]
    i_type = list(d[i].keys())[0]
    for j in range(lower,upper):
      try:
        value = valor_str(ws[j][1])
        name = ws[j][0].value
        if i_type in ['Ações','Fundos Imobiliários']:
          qtd = valor_str(ws[j][6])
        elif i_type in ['Tesouro Direto']:
          qtd = valor_str(ws[j][4])
        else:
          qtd = 0
        df_ = pd.DataFrame([[i_type,name,qtd,value]],columns =['tipo','ativo','quantity','valor'])
        df = pd.concat([df,df_])
      except:
        pass
  inv = df.reset_index(drop=True)
  return inv
```

```python
# rename all files
import warnings
warnings.filterwarnings('ignore')

df0 = pd.DataFrame()
for f in spreadsheets:
    wb = load_workbook(path  +'/' + f );
    ws = wb.active
    date_ = re.findall("\d+/\d+/\d+", ws['F1'].value)[0]
    # convert dd/mm/yyyy to yyyy-mm-dd
    date_ = date_[6:] + '-' + date_[3:5] + '-' + date_[:2]
    print(f'{f} : {date_}')
    # rename file
    os.rename(path  +'/' + f, path  +'/' + date_ + ' XpPosição.xlsx')

    wb = load_workbook(path  +'/' + date_ + ' XpPosição.xlsx');
    ws = wb.active
    df_ = read_xlsx_xp_df(ws)
    df_['date'] = date_
    df0 = pd.concat([df0, df_], axis=0)

df0['date'] = pd.to_datetime(df0['date'], format='%Y-%m-%d')
df0['date_str'] = df0['date'].dt.strftime('%Y-%m-%d')

warnings.filterwarnings('default')
```

```python
# randomizing dataframe before sharing (comment this line if you want to keep the same dataframe)
df0 = df0.sample(frac=0.5).reset_index(drop=True)

#  ussing random names
import random
import string

def randomString(stringLength=4):
    letters = string.ascii_lowercase
    return ''.join(random.choice(letters) for i in range(stringLength))

df0['ativo'] = df0['ativo'].apply(lambda x: randomString(10))

# using random values for quantity and valor
df0['quantity'] = df0['quantity'].apply(lambda x: random.randint(1, 100))
df0['valor'] = df0['valor'].apply(lambda x: random.randint(1, 1000))
```

```python
# get stock prices 

from io import StringIO
import requests
import pandas as pd

def get_stock_price(ticker = "BBAS3", date = None):
    chave_api = 'dQ5YYMCD3OJFKOSN5'
    url = f'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY_ADJUSTED&symbol=IVVB11&outputsize=full&symbol={ticker}.SAO&apikey={chave_api}&datatype=csv'
    r = requests.get(url)
    tabela = pd.read_csv(StringIO(r.text))
    if date == None:
        return tabela
    else:
        max_date = tabela.query('timestamp <= @date')['timestamp'].max()
        return tabela[tabela['timestamp'] == max_date]
```

```python
prices = pd.DataFrame()
for i in range(len(stocks_fii_list)):
    prices_ = get_stock_price(stocks_fii_list[i])
    prices_['ativo'] = stocks_fii_list[i]
    if prices_.size > 10:
        prices = pd.concat([prices,prices_],axis=0)
        print(f'{stocks_fii_list[i]}: {prices_.size}')
        time.sleep(15)
    else:
        print(f'error: {stocks_fii_list[i]}')
```

```python
def calculate_return(prices, ticker, date, how_many_days = 30, print_ = False, mock = False):
    # using mock = True to randomize the return, use mock = False to get the real return to your data
    prices = prices.sort_values(by=['timestamp'],ascending=False)
    try:
        how_many_days_adjusted = int(how_many_days * 5 / 7)
        final_date_adjusted = prices.query('ativo == @ticker').query('timestamp <= @date')['timestamp'].max()
        final_price = prices.query('ativo == @ticker').query('timestamp == @final_date_adjusted')['adjusted_close'].values[0]

        initial_date_adjusted = prices.query('ativo == @ticker').query('timestamp <= @final_date_adjusted')['timestamp'].iloc[how_many_days_adjusted]
        initial_price = prices.query('ativo == @ticker').query('timestamp == @initial_date_adjusted')['adjusted_close'].values[0]
        return_calculated = (final_price/initial_price - 1) * 100
    except:
        return_calculated = np.nan
        print(f'error: {ticker}')
    if print_:
        print(f'initial_date_adjusted: {initial_date_adjusted}')
        print(f'final_date_adjusted: {final_date_adjusted}')
        print(f'initial_price: {initial_price}')
        print(f'final_price: {final_price}')
        print(f'how_many_days_adjusted: {how_many_days_adjusted}')
    if mock:
        return np.random.normal(0,.1) * 100
    return round(return_calculated,1) 
```

```python
fii_stocks = df0[df0['tipo'].isin(['Ações','Fundos Imobiliários'])].copy()
# using mock = True to randomize the return, use mock = False to get the real return to your data
fii_stocks['return'] = fii_stocks.apply(lambda x: calculate_return(prices, x['ativo'], x['date_str'], 30,print_= False, mock = True) if x['tipo'] in ['Ações','Fundos Imobiliários'] else 0, axis=1)
last_date = fii_stocks['date'].max()
fii_stocks.query('date == @last_date').sort_values(by=['tipo','return'],ascending=False)[['ativo','tipo','return','valor']]
fii_stocks['valor_ex'] = (fii_stocks['valor'] / (1 + fii_stocks['return']/100) ).round(2)
fii_stocks['money_return'] = fii_stocks['valor'] - fii_stocks['valor_ex'] 
```
